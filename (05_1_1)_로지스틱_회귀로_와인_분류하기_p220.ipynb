```python
import pandas as pd
#6497개의 wine sample data
wine = pd.read_csv('https://raw.githubusercontent.com/youngchanseo/Hon.Gong.Machine-SNU/main/(05-1-0)%20Raw%20data.csv')
# wine = pd.read_csv('https://bit.ly/wine_csv_data')
wine.head() # 처음 5개의 sample
```

처음 3개의 열 (alchol, sugar, pH은 각각 알코올 도수, 당도, pH 값을 나타냄

네번째 열(class)은 target값으로 0이면 red wine, 1이면 white wine. 이진분류 문제이고 white wine이 양성클래스임


```python
# info() method: dtaframe의 각 열의 data type과 누락된 데이터가 있는지 확인
wine.info()
```

```python
# describe() method: 열에 대한 간략한 통계 출력
wine.describe()
```

```python
# StandardScaler class를 사용해 특성을 표준화하기전, pandas data frame을 numpy로 바꾸고 훈련세트오 ㅏ 테스트 세트로 나눔
data = wine[['alcohol', 'sugar', 'pH']].to_numpy()
target = wine['class'].to_numpy()
#wine data frame에서 처음 3개의 열을 numpy 배열로 바꿔서 data 배열에 저장하고 마지막 class열을 numpy 배열로 바꿔서 target 배열에 저장
```

```python
# 훈련세트와 테스트 세트로 나누어

from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2, random_state=42)
```

train_test_split() 함수는 설정값을 지정하지 않으면 25%를 테스트 세트로 지정함 샘플개수가 충분히 많으므로 20%정도만 테스트 세트로 나눔(test_size = 0.2)


```python
#만들어진 훈련 세트와 테스트 세트의 크기를 확인
#훈련세트는 5197개, 테스트 세트는 1300개
print(train_input.shape, test_input.shape)

#훈련세트를 전처리하고, 같은 객체를 그대로 사용해 테스트 세트를 변환
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
# 표준점수로 변환된 train_scaled와 test_scaled를 사용해 로지스틱 회귀 모델을 훈련

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_scaled, train_target)
print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```

훈련세트와 테스트 세트의 점수가 모두 낮아서 모델이 과소적합됨

=> 규제 매개변수 C의 값을 바꾸거나 solver 매개변수에서 다른 알고리즘을 선택할 수 있음 or 다항 특성을 만듦




# 설명하기 쉬운 모델과 어려운 모델


```python
#logsitic regression이 학습한 계수와 절편을 출력
print(lr.coef_, lr.intercept_)
```
