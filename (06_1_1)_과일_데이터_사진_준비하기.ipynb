```python
import numpy as np
import matplotlib.pyplot as plt
fruits = np.load('fruits_300.npy') # load() method에 파일 이름을 전달함: numpy에서 npy 파일 로드
print(fruits.shape)
```

이 배열에서

첫번째 차원(300): 샘플의 개수

두 번째 차원 (100)은 이미지 높이,

세번째 차원(100): 이미지 너비

```python
print(fruits[0, 0,:])
```

첫 번째 행에 있는 픽셀 100개에 들어있는 값을 출력함.

이 numpy 배열은 흑백 사진을 담고 있으므로, 0 ~ 255까지의 정숫값을 가짐

```python
plt.imshow(fruits[0], cmap='gray')
plt.show()
```

```python
#cmap 매개변수를 'gray_r'로 지정: 반전
plt.imshow(fruits[0],cmap='gray_r')
plt.show()
```

이 그림에서 밝은 부분이 0에 가깝고, 짙은 부분이 255에 가까운 값

```python
#이 데이터는 사과, 바나나, 파인애플이 각각 100개씩 들어있음
fig, axs =plt.subplots(1,2)
axs[0].imshow(fruits[100], cmap='gray_r')
axs[1].imshow(fruits[200], cmap='gray_r')
plt.show()

# axs[0]에는 파인애플 이미지를, axs[1] 에는 바난 이미지를 그렸음
```
